# Content Based Filtering


Now I will experiment with description data, looking at relationships based on what the our sample users liked watching and reccommend other titles to them

## Load Data

```python
movies = pd.read_csv('data/movies_small.csv', sep = ';')

movies.head()
```

## Vectorize Descriptions

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

# handling missing data
movies['overview'] = movies['overview'].fillna('')

movies['overview']
```

```python
# Matrix

tfidf_matrix = tfidf.fit_transform(movies['overview'])

tfidf_matrix
```

```python
pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())

# values represent how rare the term appears in the data set as well how unique it is to a specific row (movie)
```

```python
tfidf_matrix.shape
```

## Similarity Matrix

```python
from sklearn.metrics.pairwise import linear_kernel

similarity_matrix = linear_kernel(tfidf_matrix, tfidf_matrix)

similarity_matrix 
```

```python
# Print the first row again
print(similarity_matrix[0])
```

## Finding Movies that have the Strongest Similarities

```python
movie_title = 'Cars 2'
```

```python
idx = movies.loc[movies['title']==movie_title].index[0]
idx
```

```python
# Enumerating into pairs in a tuple list to keep track of movie title
scores = list(enumerate(similarity_matrix[idx]))
scores
```

```python
# sorting by scores
scores = sorted(scores, key=lambda x: x[1], reverse = True)
scores
```

```python
# retrieving indices of movie titles to use as pointers
movie_indices = [tpl[0] for tpl in scores[1:4]]
movie_indices
```

```python
# Find title(s) based on indices

title = list(movies['title'].iloc[movie_indices])
title
```

```python
# Putting everything together

def similiarity_movies(movie_title, num_movies):
    idx = movies.loc[movies['title']==movie_title].index[0]
    scores = list(enumerate(similarity_matrix[idx]))
    
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    movie_indices = [tpl[0] for tpl in scores[1:num_movies + 1]]

    title = list(movies['title'].iloc[movie_indices])

    return title
```
